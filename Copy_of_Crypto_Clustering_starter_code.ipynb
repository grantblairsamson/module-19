```python
# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import hvplot.pandas
# Import holoviews and enable the bokeh extension
import holoviews as hv
hv.extension('bokeh')
```

```python
# Upload and load the data into a DataFrame
from google.colab import files
uploaded = files.upload()

# Load the crypto market data CSV file
crypto_df = pd.read_csv("crypto_market_data.csv")
crypto_df.head()  # Display the first few rows of the DataFrame
```

```python
# Display the summary statistics of the data
crypto_df.describe()
```

```python
# Scale the data using StandardScaler
scaler = StandardScaler()
crypto_scaled = scaler.fit_transform(crypto_df.drop(columns=["coin_id"]))

# Create a DataFrame with the scaled data
scaled_df = pd.DataFrame(crypto_scaled, columns=crypto_df.columns[1:], index=crypto_df["coin_id"])
scaled_df.head()
```

```python
# Elbow method to find the best value for k
inertia = []
k_values = range(1, 12)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_df)
    inertia.append(kmeans.inertia_)

# Create a DataFrame to plot the Elbow Curve
elbow_data = {"k": k_values, "inertia": inertia}
elbow_df = pd.DataFrame(elbow_data)

# Initialize hvplot extension
hvplot.extension("bokeh")

# Plot the Elbow Curve using the recommended format
elbow_df.hvplot.line(
    x="k",
    y="inertia",
    width=1200,
    height=400,
    title="Elbow Curve",
    xlabel="k",
    ylabel="Inertia",
    rot=90
)
```

```python
# Based on the Elbow Curve, let's choose the best value for k (typically at the 'elbow')
best_k = 4  # Update based on your visual analysis from the Elbow Curve
```

```python
# Initialize and fit the K-means model
kmeans_model = KMeans(n_clusters=best_k, random_state=42, n_init=10)
crypto_clusters = kmeans_model.fit_predict(scaled_df)

# Add predicted clusters to the original DataFrame
crypto_df["Cluster"] = crypto_clusters
crypto_df.head()  # Display the first few rows with clusters
```

```python
# Ensure hvplot extension is initialized
hvplot.extension("bokeh")

# Check if the DataFrame has valid cluster assignments and the necessary columns
print(crypto_df[["price_change_percentage_24h", "price_change_percentage_7d", "Cluster"]].head())

# Scatter plot of the clusters
scatter_plot = crypto_df.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="Cluster",
    hover_cols=["coin_id"],
    width=1200,
    height=400,
    title="Cluster Scatter Plot",
    xlabel="24h Price Change",
    ylabel="7d Price Change"
)

# Display the scatter plot inline
scatter_plot
```

```python
# Perform PCA to reduce the number of features to 3 components
pca = PCA(n_components=3)
crypto_pca = pca.fit_transform(scaled_df)

# Create a DataFrame for the PCA results
pca_df = pd.DataFrame(crypto_pca, columns=["PC1", "PC2", "PC3"], index=crypto_df["coin_id"])
pca_df.head()  # Display the first few rows of the PCA-transformed data
```

```python
# Display the total explained variance of the three principal components
explained_variance = pca.explained_variance_ratio_.sum()
print(f"Total Explained Variance: {explained_variance:.2%}")
```

```python
# Ensure hvplot extension is initialized
hvplot.extension("bokeh")

# Elbow method using PCA data to find the best value for k
inertia_pca = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(pca_df)
    inertia_pca.append(kmeans.inertia_)

# Create a DataFrame to plot the Elbow Curve for PCA data
elbow_pca_data = {"k": k_values, "inertia": inertia_pca}
elbow_pca_df = pd.DataFrame(elbow_pca_data)

# Print the elbow_pca_df to verify the data
print(elbow_pca_df)

# Plot the Elbow Curve for PCA data using hvplot
elbow_pca_plot = elbow_pca_df.hvplot.line(
    x="k",
    y="inertia",
    width=1200,
    height=400,
    title="Elbow Curve for PCA Data",
    xlabel="Number of Clusters (k)",
    ylabel="Inertia",
    rot=0  # Make sure labels don't rotate unnecessarily
)

# Display the Elbow Curve for PCA data
elbow_pca_plot
```

```python
# Choose the best value for k based on the Elbow Curve for PCA data
best_k_pca = 4  # Update based on your visual analysis
```

```python
# Perform K-Means clustering on the PCA data
kmeans_pca = KMeans(n_clusters=best_k_pca, random_state=42, n_init=10)
pca_clusters = kmeans_pca.fit_predict(pca_df)

# Add the cluster assignments to the PCA DataFrame
pca_df["Cluster"] = pca_clusters
pca_df.head()  # Display the first few rows with clusters
```

```python
# Ensure hvplot extension is initialized
hvplot.extension("bokeh")

# Check if the DataFrame has valid PCA components and cluster assignments
print(pca_df[["PC1", "PC2", "Cluster"]].head())  # Ensure that PC1, PC2, and Cluster exist and have valid data

# Scatter plot for clusters based on PCA data
pca_scatter_plot = pca_df.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="Cluster",
    hover_cols=["coin_id"],
    width=1200,
    height=400,
    title="PCA Cluster Scatter Plot",
    xlabel="PC1",
    ylabel="PC2"
)

# Display the scatter plot inline
pca_scatter_plot
```

```python
# Ensure hvplot extension is initialized
hvplot.extension("bokeh")

# Compare the Elbow Curves (Original vs PCA)
elbow_compare = (
    elbow_df.hvplot.line(x="k", y="inertia", title="Elbow Curve for Original Data", width=400, height=400) +
    elbow_pca_df.hvplot.line(x="k", y="inertia", title="Elbow Curve for PCA Data", width=400, height=400)
)

# Display the elbow comparison plot
elbow_compare_plot = elbow_compare
elbow_compare_plot

# Compare the cluster scatter plots (Original vs PCA)
cluster_compare = (
    crypto_df.hvplot.scatter(
        x="price_change_percentage_24h",
        y="price_change_percentage_7d",
        by="Cluster",
        hover_cols=["coin_id"],
        title="Original Data Clusters",
        width=1000,
        height=400
    ) +
    pca_df.hvplot.scatter(
        x="PC1",
        y="PC2",
        by="Cluster",
        hover_cols=["coin_id"],
        title="PCA Data Clusters",
        width=1000,
        height=400
    )
)

# Display the cluster comparison plot
cluster_compare_plot = cluster_compare
cluster_compare_plot
```
